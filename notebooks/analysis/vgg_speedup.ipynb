```python
speedup_df = pd.read_csv('vgg_speedup.csv',sep=r'\s*,\s*')
```

```python
sb.pointplot(x="GPUs", y="Speedup", hue="Schedule", data=speedup_df )
plt.savefig('../../Figures/VGG_CIFAR_SPEEDUP.png')
```

```python
sb.pointplot(x="GPUs", y="Effeciency", hue="Schedule", data=speedup_df )
plt.savefig('../../Figures/VGG_CIFAR_EFF.png')
```

```python
sb.pointplot(x="GPUs", y="Speedup", hue="Threads", data=speedup_df )
plt.savefig('../../Figures/ResNet_CIFAR_SPEEDUP.png')
```

```python
sb.pointplot(x="GPUs", y="Effeciency", hue="Threads", data=speedup_df )
plt.savefig('../../Figures/ResNet_CIFAR_EFF.png')
```
